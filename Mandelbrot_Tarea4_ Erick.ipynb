```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Crear una malla de puntos (a, b)
a, b = np.meshgrid(np.linspace(-1.5, 1, 100), np.linspace(-1, 1, 100))

# Definir la constante compleja c = a + b*i
c = a + b * 1j

# Inicializar z
z = 0

# Umbral para la condicin de terminacin
umbral = 100

# Iteracin de la funcin
for c1 in range(1, 101):
    z = z**2 + c
    mag_z = abs(z)  # Calcular el valor absoluto de z

    # Colorear cada iteracin con un color diferente
    color = plt.cm.cool(c1/101)  # Escala de colores entre 0 y 1
    plt.scatter(a[mag_z < umbral], b[mag_z < umbral], color=color, s=2)

# Mostrar el grfico resultante
plt.show()
```
