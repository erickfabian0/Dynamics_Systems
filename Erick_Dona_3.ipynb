```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, transform
```

```python

img = color.rgb2gray(transform.resize(io.imread('LSD.jpg'), (100, 100)))
x, y = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))

xa = np.array([x.flatten(), y.flatten()])
A = np.array([[2, 1], [1, 1]])
xb = np.mod(A @ xa, 1)

plt.subplot(1, 2, 1)
plt.scatter(xa[0], xa[1], c=img.flatten(), marker='o', cmap='gray')
plt.title('Original')

plt.subplot(1, 2, 2)
plt.scatter(xb[0], xb[1], c=img.flatten(), marker='o', cmap='gray')
plt.title('Transformada')

x = xb[0]
y = xb[1]
yt = (1 + 0.5 * np.cos(2 * np.pi * y)) * np.cos(2 * np.pi * x)
xt = (1 + 0.5 * np.cos(2 * np.pi * y)) * np.sin(2 * np.pi * x)
zt = 0.5 * np.sin(2 * np.pi * y)

plt.figure()
plt.scatter(xt, yt, zt, c=img.flatten(), marker='o', cmap='gray')
plt.axis('equal')
plt.title('Gráfica 3D')
plt.show()
```
