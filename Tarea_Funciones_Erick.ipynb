```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
# Definir la función f(x) = sin(x)
def f(x):
    return np.sin(x)

# Definir el valor inicial x = 1
x = .9

# Definir el número de iteraciones n
n = 20

# Crear una lista vacía para guardar la órbita
orbita = []

# Usar un bucle for para iterar la función f
plt.figure()
for i in range(n):
    # Aplicar la función f al valor x
    y = f(x)
    plt.plot([x, x], [x, y], 'k')
    plt.plot([x, y], [y, y], 'k')
    x = y
    # Añadir el resultado al vector órbita
    orbita.append(x)

    # Mostrar las coordenadas de la iteración actual
    coords = f"({x:.3f}, {y:.3f})"
    plt.annotate(coords, (x, y), textcoords="offset points", xytext=(5, -10), ha='center', fontsize=8)

# Mostrar el vector órbita
print(orbita)

# Convertir el vector órbita a un arreglo numpy
orbita_arr = np.array(orbita)

# Encontrar los valores mínimo y máximo de la órbita
xmin = np.min(orbita_arr)
xmax = np.max(orbita_arr)

# Crear un arreglo de valores para x
x_vals = np.linspace(xmin, xmax, 10)

# Graficar la función f(x) = sin(x) y la línea identidad y = x
plt.plot(x_vals, x_vals, label='y = x', linewidth=2)
plt.plot(x_vals, f(x_vals), label='f(x) = sin(x)', linewidth=2)

# Encontrar el punto de intersección entre la recta tangente y la función sin(x)
intercept_x = []
intercept_y = []
for i in range(len(orbita)-1):
    if np.isclose(orbita[i], orbita[i+1]):
        intercept_x.append(orbita[i])
        intercept_y.append(f(orbita[i]))

# Graficar el punto de intersección
plt.scatter(intercept_x, intercept_y, color='red', label='Intersección')

# Mostrar las coordenadas de los puntos de intersección
for i in range(len(intercept_x)):
    coords = f"({intercept_x[i]:.3f}, {intercept_y[i]:.3f})"
    plt.annotate(coords, (intercept_x[i], intercept_y[i]), textcoords="offset points", xytext=(-10,-10), ha='center', fontsize=8)

# Configurar el aspecto del gráfico
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Iteración de la función f(x) = sin(x)')
plt.legend()

# Mostrar el gráfico resultante
plt.show()
```
